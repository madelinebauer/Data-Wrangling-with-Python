## 3a. Data Wrangling with Python: Activity 1 page 17

```python
# Create a list of 100 random numbers
LIMIT = 100
list_1 = [random.randint(0, LIMIT) for x in range(0, LIMIT)]
```

```python
print(list_1)
```

```python
# Create a new list from this random list, with numbers that are divisible by 3.
divByThree = [i for i in list_1 if i % 3 == 0]
```

```python
print(divByThree)
```

```python
# Calculate the length of these two lists and store the difference in a new variable
length_list1 = len(list_1)
length_divByThree = len(divByThree)
difference = length_list1 - length_divByThree
```

```python
print(difference)
```

```python
# Using a loop, perform steps 2 and 3 and find the difference varialbe three times.
repeat = 3
diff_list = []
for i in range(0, repeat):
    list_1 = [random.randint(0, LIMIT) for x in range(0, LIMIT)]
    divByThree = [i for i in list_1 if i % 3 == 0]
    length_list1 = len(list_1)
    length_divByThree = len(divByThree)
    difference = length_list1 - length_divByThree
    
    diff_list.append(difference)
diff_list
```

```python
# Find the arithmetic mean of these three difference values.
ari_mean = sum(diff_list) / float(len(diff_list))
ari_mean
```

```python
# Find the type and length of the multiline_text string
type(multiline_text)
```

```python
len(multiline_text)
```

```python
# Remove all new lines and symbols using the replace function
```

```python
multiline_text = multiline_text.replace('\n', "")
multiline_text
```

```python
# Remove all special characters
clean_text = ""
for char in multiline_text:
    if char == " ":
        clean_text += char
    elif char.isalnum():
        clean_text += char
    else:
        clean_text += " "
```

```python
clean_text
```

```python
# Find all of the words in multiline_text using the split function
```

```python
word_list = clean_text.split()
word_list
```

```python
# Create a list from this list that will contain only the unique words.
```

```python
unique_words = list(set(word_list))
len(unique_words)
```

```python
# Create dictionary for unique words
unique_words_dict = dict.fromkeys(word_list)
len(list(unique_words_dict.keys()))
```

```python
# Count of times the unique word has appeared in the list using the key and value in dict.
for word in word_list:
    if unique_words_dict[word] is None:
        unique_words_dict[word] = 1
    else:
        unique_words_dict[word] += 1

unique_words_dict
```

```python
# Find the top 25 words from the unique words that you have found using the slice function.
```

```python
top_25 = sorted(unique_words_dict.items(), reverse = True)
```

```python
top_25 # sorted alphabetically on accident
```

```python
# sort by value instead of key
top_25 = sorted(unique_words_dict.items(), key = lambda key_val_tuple:
               key_val_tuple[1], reverse = True)
top_25[:25]
```
